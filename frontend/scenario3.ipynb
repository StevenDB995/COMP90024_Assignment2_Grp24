```python

# total number of posts with domestic violence each year
# Normalize tags and filter posts with 'domesticviolence'
DVData_df['tags'] = DVData_df['tags'].apply(lambda x: [tag.lower() for tag in x])
filtered_posts = DVData_df[DVData_df['tags'].apply(lambda x: 'domesticviolence' in x)]

# Extract year from 'created_at'
filtered_posts['year'] = pd.to_datetime(filtered_posts['created_at']).dt.year

# Aggregate counts by year
yearly_counts = filtered_posts['year'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
yearly_counts.plot(kind='bar', color='cyan')
plt.title('Number of Domestic Violence Related Posts Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

#Graph of the total number of people affected by domestic violence each year
# Identify rate columns correctly
rate_columns = [col for col in AffectFamilyMem_df.columns if "affected_family_members_rate_per_100k_" in col]

# Check column extraction
print("Rate columns:", rate_columns)

# Sum rates per year based on the identified columns
sums_per_year = AffectFamilyMem_df[rate_columns].sum()

# Extract years from the rate column names and map them to sums
years = [col.split('_')[-2] for col in rate_columns]
yearly_totals = pd.Series(data=sums_per_year.values, index=years)

# Sorting the years for chronological plotting
yearly_totals.sort_index(inplace=True)

# Plotting the aggregated data
plt.figure(figsize=(10, 6))
yearly_totals.plot(kind='bar', color='magenta')
plt.title('Total Affected Family Members Rates per 100k Each Year')
plt.xlabel('Year')
plt.ylabel('Sum of Rates per 100k')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#predicting the total number for futher predicting
# Extract rate columns and calculate sum per year
rate_columns = [col for col in AffectFamilyMem_df.columns if "affected_family_members_rate_per_100k_" in col]
sums_per_year = AffectFamilyMem_df[rate_columns].sum()

# Prepare data for linear regression
years = np.arange(2013, 2018)
X = years.reshape(-1, 1)
y = sums_per_year.values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict future years
future_years = np.arange(2018, 2025).reshape(-1, 1)
predictions = model.predict(future_years)

# Combine historical and predicted data
all_years = np.concatenate([years, future_years.ravel()])
all_sums = np.concatenate([y, predictions])

# Plot the results with the predicted points connected by a dashed line
plt.figure(figsize=(10, 6))
plt.scatter(years, y, color='blue', label='Historical Data')
plt.plot(years, y, color='blue', linestyle='-', marker='o')
plt.scatter(future_years.ravel(), predictions, color='red', label='Predicted Data')
plt.plot(future_years.ravel(), predictions, color='red', linestyle='--', marker='o')
plt.title('Total Affected Family Members per Year (2013-2024)')
plt.xlabel('Year')
plt.ylabel('Total Affected Family Members')
plt.xticks(np.arange(2013, 2025))
plt.legend()
plt.grid(True)

# Annotate the plot with all values
for i, total in enumerate(y):
    plt.annotate(f"{total:.2f}", (years[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
for i, total in enumerate(predictions):
    plt.annotate(f"{total:.2f}", (future_years.ravel()[i], predictions[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
#Ratio of predicted affected family members to number of posts
# Normalize tags and filter posts with 'domesticviolence'
DVData_df['tags'] = DVData_df['tags'].apply(lambda x: [tag.lower() for tag in x])
filtered_posts = DVData_df[DVData_df['tags'].apply(lambda x: 'domesticviolence' in x)]

# Extract year from 'created_at'
filtered_posts['year'] = pd.to_datetime(filtered_posts['created_at']).dt.year

# Aggregate counts by year
yearly_counts = filtered_posts['year'].value_counts().sort_index()
# Extracting the number of posts for 2023 and 2024
posts_2023 = yearly_counts[2023]
posts_2024 = yearly_counts[2024]
# Extracting the predicted values 
predicted_2018 = predictions[future_years.ravel() == 2018][0]
predicted_2019 = predictions[future_years.ravel() == 2019][0]
predicted_2020 = predictions[future_years.ravel() == 2020][0]
predicted_2021 = predictions[future_years.ravel() == 2021][0]
predicted_2022 = predictions[future_years.ravel() == 2022][0]
predicted_2023 = predictions[future_years.ravel() == 2023][0]
predicted_2024 = predictions[future_years.ravel() == 2024][0]

# Calculating the ratios
ratio_2023 = predicted_2023 / posts_2023
ratio_2024 = predicted_2024 / posts_2024

# Creating a DataFrame for plotting
ratio_df = pd.DataFrame({
    'Year': [2023, 2024],
    'Ratio': [ratio_2023, ratio_2024]
})

# Plotting the ratio as a line chart
plt.figure(figsize=(8, 5))
plt.plot(ratio_df['Year'], ratio_df['Ratio'], marker='o', linestyle='-', color='blue')
plt.title('Ratio of Predicted Affected Family Members to Number of Posts (2023-2024)')
plt.xlabel('Year')
plt.ylabel('Ratio (Predicted Affected Family Members / Number of Posts)')
plt.xticks([2023, 2024])
plt.grid(True)

# Annotate the plot with the ratio values
for i, ratio in enumerate(ratio_df['Ratio']):
    plt.annotate(f"{ratio:.2f}", (ratio_df['Year'][i], ratio_df['Ratio'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
#Growth Rate of Posts and Engagement (2023-2024)
# Extract year from dates
DVData_df['year'] = pd.to_datetime(DVData_df['created_at']).dt.year

# Calculate the number of posts per year
annual_posts = DVData_df.groupby('year').size().reset_index(name='total_posts')
annual_posts.loc[annual_posts['year'] == 2024, 'total_posts'] *= 2

# Calculate total engagements (favourites, reblogs, and replies) per year
DVData_df['total_engagement'] = DVData_df['favourites_count'] + DVData_df['reblogs_count'] + DVData_df['replies_count']
annual_engagement= DVData_df.groupby('year')['total_engagement'].sum().reset_index()
annual_engagement.loc[annual_engagement['year'] == 2024, 'total_engagement'] *= 2

# Calculate growth rates
annual_posts['growth_rate'] = annual_posts['total_posts'].pct_change() * 100
annual_engagement['growth_rate'] = annual_engagement['total_engagement'].pct_change() * 100

# Filter data for only 2024
growth_rate_2024 = {
    'year': [2024],
    'posts_growth_rate': annual_posts[annual_posts['year'] == 2024]['growth_rate'].values,
    'engagement_growth_rate': annual_engagement[annual_engagement['year'] == 2024]['growth_rate'].values
}
growth_rate_2024_df = pd.DataFrame(growth_rate_2024)

# Plot the line chart
plt.figure(figsize=(6, 4))
plt.plot(growth_rate_2024_df['year'], growth_rate_2024_df['posts_growth_rate'], marker='o', linestyle='-', color='blue', label='Posts Growth Rate')
plt.plot(growth_rate_2024_df['year'], growth_rate_2024_df['engagement_growth_rate'], marker='o', linestyle='-', color='green', label='Engagement Growth Rate')
plt.title('Growth Rate of Posts and Engagement (2024)')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.xticks([2024])  # Ensure x-axis only shows 2024
plt.legend()
plt.grid(True)

# Annotate the plot with data labels
for i in range(len(growth_rate_2024_df)):
    plt.annotate(f"{growth_rate_2024_df['posts_growth_rate'][i]:.2f}%", (growth_rate_2024_df['year'][i], growth_rate_2024_df['posts_growth_rate'][i]), textcoords="offset points", xytext=(0,10), ha='center', color='blue')
    plt.annotate(f"{growth_rate_2024_df['engagement_growth_rate'][i]:.2f}%", (growth_rate_2024_df['year'][i], growth_rate_2024_df['engagement_growth_rate'][i]), textcoords="offset points", xytext=(0,-15), ha='center', color='green')

plt.show()
annual_engagement
```
